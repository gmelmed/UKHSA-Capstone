```python
from docling.document_converter import DocumentConverter
converter = DocumentConverter()
```

```python
source = "ethiopia_nap.pdf"

result = converter.convert(source)
```

```python
print(result.document.export_to_markdown())
```

```python
import pymupdf4llm
```

```python
chile_pdf = "chile_nap.pdf"
```

```python
chile_text = pymupdf4llm.to_markdown(chile_pdf)

with open("chile_nap.md", "w") as f:
    f.write(chile_text)
```

```python
with open("chile_nap.txt", "w") as f:
    f.write(chile_text)
```
